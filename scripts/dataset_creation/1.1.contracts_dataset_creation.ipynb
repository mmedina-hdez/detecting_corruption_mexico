## submission_deadline_date

```python
contratos2010to2017['submission_deadline_date'].isnull().sum() / len(contratos2010to2017)
```

```python
#For the 2010 to 2017 dataset
#split the dataset in two, one with the contracts that in their column opening date have the characters ':00 GMT', and those who dont have it
contratos2010to2017_1 = contratos2010to2017[contratos2010to2017['submission_deadline_date'].str.contains(':00 GMT', na=False)]
contratos2010to2017_2 = contratos2010to2017[~contratos2010to2017['submission_deadline_date'].str.contains(':00 GMT', na=False)]

```

```python
#SPLIT1
#search in column opening date the string ':00 GMT' pattern and replace it with ''
contratos2010to2017_1['submission_deadline_date'] = contratos2010to2017_1['submission_deadline_date'].str.replace(' GMT', '')
#keep the first 10 characters of the string
contratos2010to2017_1['submission_deadline_date'] = contratos2010to2017_1['submission_deadline_date'].str[:10]
#without specifying , I will only erase the hour and minutes
contratos2010to2017_1['submission_deadline_date'] = pd.to_datetime(contratos2010to2017_1['submission_deadline_date'])
```

```python
#SPLIT2
#keep the first 10 characters of the string
contratos2010to2017_2['submission_deadline_date'] = contratos2010to2017_2['submission_deadline_date'].str[:10]
#for the second split ill do the same
contratos2010to2017_2['submission_deadline_date'] = pd.to_datetime(contratos2010to2017_2['submission_deadline_date'], format='%d/%m/%Y', errors='raise')
```

```python
#reunify the splits
contratos2010to2017 = pd.concat([contratos2010to2017_1, contratos2010to2017_2], ignore_index=True)
```
